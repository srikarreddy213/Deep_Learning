```python
#1  Import
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dataset
X = np.array([
    [1,1],  # E1
    [1,0],  # E2
    [0,1],  # E3
    [0,0]   # E4
])

y = np.array([1,1,1,0])  # OR logic

X, y
```

```python
#Initialize
w = np.zeros(2)
b = 0.0
learning_rate = 1
epochs = 10
```

```python
# Activation Function
def activation_function(z):
    if z > 0:
        return 1
    else:
        return 0
```

```python
# Training
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = activation_function(z)

        error = y[i] - y_pred

        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

print("Final Weights:", w)
print("Final Bias:", b)
```

```python
#Testing
print("\nTesting Email Classifier")

for i in range(len(X)):
    z = np.dot(w, X[i]) + b
    prediction = activation_function(z)
    print("Email:", X[i], "Prediction:", prediction)
```

```python
#Decision Boundary Plot
plt.scatter(X[:,0], X[:,1], c=y)

x_vals = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    y_vals = -(w[0]*x_vals + b) / w[1]
    plt.plot(x_vals, y_vals)

plt.xlabel("Suspicious Keywords")
plt.ylabel("Unknown Sender")
plt.title("Spam Detection (OR Logic)")
plt.grid()
plt.show()
```
