Perceptron (Binary Classification)
- Example : Study Hours :

Step-1:Prepare Training Data

```python
import numpy as np
```

```python
#Training Data
X=np.array([1,2,3,4]) #Study Hours
y=np.array([0,0,1,1]) #Fail(0)/Pass(1)
X,y
```

```python
from typing import LiteralString
w=0.0
b=0.0
learning_rate=1
epochs=5
print("Initial Weight:",w)
print("Initial Bias:",b)
```

# **Step 3: Define Activation Function (Step Function)**

```python
def activation_function(x):
  if x>0:
    return 1
  else:
    return 0
```

# **Training Loop**

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):

        # 1. Compute weighted sum
        z = w * X[i] + b

        # 2. Apply activation
        y_pred = activation_function(z)

        # 3. Compute error
        error = y[i] - y_pred

        # 4. Update weight
        w = w + learning_rate * error * X[i]

        # 5. Update bias
        b = b + learning_rate * error

        print(f"Input:{X[i]}, Predicted:{y_pred}, Actual:{y[i]}, w:{w}, b:{b}")
```

# Final Model Parameters

```python
print("\nFinal Weight:", w)
print("Final Bias:", b)
```

# **Testing The Model**

```python
test_value = 2.9
z = w * test_value + b
prediction = activation_function(z)

print(f"\nFor study hours {test_value}, Prediction:", prediction)
```

```python
import numpy as np
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
    ])
y=np.array([0,0,0,1]) #AND gate output
X,y
```

```python
w=np.zeros(2)
b=0.0
learning_rate=1
epochs=10
print("Initial Weights:",w)
print("Initial Bias:",b)
```

```python
# Activation function
def activation_function(z):
    if z > 0:
        return 1
    else:
        return 0
```

```python
# Training
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):

        # weighted sum
        z = np.dot(w, X[i]) + b

        # prediction
        y_pred = activation_function(z)

        # error
        error = y[i] - y_pred

        # update
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

        print(f"Input:{X[i]}, Pred:{y_pred}, Actual:{y[i]}, w:{w}, b:{b}")
```

```python
print("\nFinal Weights:", w)
print("Final Bias:", b)
```

```python
print("\nTesting AND Gate")

for i in range(len(X)):
    z = np.dot(w, X[i]) + b
    prediction = activation_function(z)  # make sure name matches
    print("Input:", X[i], "Prediction:", prediction)
```

```python
import matplotlib.pyplot as plt

# plot points
plt.scatter(X[:,0], X[:,1], c=y)

# decision boundary line
x_vals = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    y_vals = -(w[0]*x_vals + b) / w[1]
    plt.plot(x_vals, y_vals)

plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.title("AND Gate Decision Boundary")
plt.grid()
plt.show()
```
