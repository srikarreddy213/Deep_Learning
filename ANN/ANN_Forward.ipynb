```python
import numpy as np
#define XOR Dataset
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[1],[1],[0]])
X,y
```

# **Define Step Activation Function**

```python
def step(z):
  return 1 if z>0 else 0
```

# **Inititalize the Weights and biases(Manually choosen)**

```python


# Hidden layer
w1, w2, b1 = 1, 1, -0.5
w3, w4, b2 = 1, 1, -1.5

# Output layer
w5, w6, b3 = 1, -2, -0.5.
print("Weights initialized successfully")
```

# **Define Forward Propagration Function**

```python
def forward_propogation(x1, x2):

    # Hidden Layer
    h1 = w1*x1 + w2*x2 + b1
    a_h1 = step(h1)

    h2 = w3*x1 + w4*x2 + b2
    a_h2 = step(h2)

    # Output Layer
    y_out = w5*a_h1 + w6*a_h2 + b3
    output = step(y_out)

    return output
```

#**Test Forward Propogration for all XOR inputs**

```python
# Test
print("XOR Results using ANN Forward Propagation")
for x in X:
    prediction = forward_propogation(x[0], x[1])
    print(f"Input: {x} Output: {prediction}")
```

# **Plotting to view the decision boundary**

```python
#plotting the decision boundary
import matplotlib.pyplot as plt
x1=np.linspace(-0.5,1.5,100)
x2=np.linspace(-0.5,1.5,100)
X1,X2=np.meshgrid(x1,x2)
Z=np.zeros_like(X1)
for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        Z[i,j]=forward_propogation(X1[i,j],X2[i,j])
plt.contourf(X1,X2,Z,alpha=0.5)
plt.scatter(X[:,0],X[:,1],c='red',edgecolors='k')
plt.title('XOR Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
